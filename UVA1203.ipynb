```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

# Classe Heap
class Heap:
  def __init__(self):
    # Inicializa uma lista vazia para representar a heap
    self.heap = []

  # Função para obter o índice do pai de um nó
  def pai(self, i):
    # O pai é (i - 1) // 2
    return (i - 1) // 2

  # Função para obter o índice do filho esquerdo de um nó
  def filho_esquerda(self, i):
    return 2 * i + 1

  # Função para obter o índice do filho direito de um nó
  def filho_direita(self, i):
    return 2 * i + 2

  # Função para inserir um valor na heap
  def inserir(self, valor):
    # .append pois é uma lista
    self.heap.append(valor)
    # Chama a função upheap para restaurar a propriedade da heap
    self.upheap(len(self.heap) - 1)

  # Função para restaurar a propriedade da heap após a inserção
  def upheap(self, i):
    # Enquanto o índice for maior que 0 e o valor do pai for maior que o valor do filho
    while i > 0 and self.heap[self.pai(i)][0] > self.heap[i][0]:
      # Troca o valor do pai com o valor do filho (REORGANIZANDO)
      self.heap[i], self.heap[self.pai(i)] = self.heap[self.pai(i)], self.heap[i]
      # Substitui
      i = self.pai(i)

  # Função para extrair o menor valor da heap
  def extrair_min(self):
    if not self.heap:
      return None
    # Se a lista heap tiver apenas 1 valor
    if len(self.heap) == 1:
      # .pop extrai e retorna
      return self.heap.pop()
    # O valor mínimo vai ser o primeiro item da lista heap
    min_valor = self.heap[0]
    # Extrai e retorna o valor do heap com .pop
    self.heap[0] = self.heap.pop()
    # Chama downheap para reorganizar
    self.downheap(0)

    return min_valor

  # Função para restaurar a propriedade da min-heap após a remoção do mínimo
  def downheap(self, i):
    while self.filho_esquerda(i) < len(self.heap):
      menor_filho_indice = self.filho_esquerda(i)
      if self.filho_direita(i) < len(self.heap) and self.heap[self.filho_direita(i)][0] < self.heap[self.filho_esquerda(i)][0]:
        menor_filho_indice = self.filho_direita(i)

      if self.heap[i][0] > self.heap[menor_filho_indice][0]:
        self.heap[i], self.heap[menor_filho_indice] = self.heap[menor_filho_indice], self.heap[i]
        i = menor_filho_indice
      else:
        break

  # Função para verificar se a heap está vazia
  def is_empty(self):
    return len(self.heap) == 0

  # Função para obter o menor valor da heap (sem remover)
  def get_min(self, next_time):
    if not self.heap:
      return None
    return self.heap[0]
```

## Design your implementation of the linked list. You can choose to use a singly or doubly linked list.

A node in a singly linked list should have two attributes: `val` and `next`. `val` is the value of the current node, and `next` is a pointer/reference to the next node.

If you want to use the doubly linked list, you will need one more attribute `prev` to indicate the previous node in the linked list. Assume all nodes in the linked list are 0-indexed.

### Implement the `MyLinkedList` class:

- `MyLinkedList()` Initializes the `MyLinkedList` object.
- `int get(int index)` Get the value of the `index`th node in the linked list. If the index is invalid, return `-1`.
- `void addAtHead(int val)` Add a node of value `val` before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
- `void addAtTail(int val)` Append a node of value `val` as the last element of the linked list.
- `void addAtIndex(int index, int val)` Add a node of value `val` before the `index`th node in the linked list. If `index` equals the length of the linked list, the node will be appended to the end of the linked list. If `index` is greater than the length, the node will not be inserted.
- `void deleteAtIndex(int index)` Delete the `index`th node in the linked list, if the index is valid.

### Example 1:

**Input**
```
["MyLinkedList", "addAtHead", "addAtTail", "addAtIndex", "get", "deleteAtIndex", "get"]
[[], [1], [3], [1, 2], [1], [1], [1]]
```

**Output**
```
[null, null, null, null, 2, null, 3]
```

**Explanation**
```
MyLinkedList myLinkedList = new MyLinkedList();
myLinkedList.addAtHead(1);
myLinkedList.addAtTail(3);
myLinkedList.addAtIndex(1, 2);    // linked list becomes 1->2->3
myLinkedList.get(1);              // return 2
myLinkedList.deleteAtIndex(1);    // now the linked list is 1->3
myLinkedList.get(1);              // return 3
```

### Constraints:

- `0 <= index, val <= 1000`
- Please do not use the built-in LinkedList library.
- At most 2000 calls will be made to `get`, `addAtHead`, `addAtTail`, `addAtIndex`, and `deleteAtIndex`.

````python
class MyLinkedList(object):

    def __init__(self):
        # Construtor da classe MyLinkedList.
        # Inicializa o atributo 'head' como None, indicando que a lista está vazia no início.
        # 'head' é uma variável que armazena o primeiro nó da lista encadeada.
        self.head = None

    def get(self, index: int) -> int:
        """
        Retorna o valor do nó no índice especificado.
        Se o índice for inválido, retorna -1.
        """

        # 'current' é uma variável temporária que percorre a lista, começando pelo nó 'head'.
        current = self.head

        # 'count' é um contador que rastreia a posição atual na lista.
        count = 0

        # Loop que percorre a lista até encontrar o nó no índice desejado ou chegar ao fim da lista.
        while current and count < index:
            # Move 'current' para o próximo nó na lista.
            current = current.next

            # Incrementa o contador.
            count += 1

        # Se 'current' não for None (ou seja, se o índice for válido), retorna o valor do nó.
        # Caso contrário, retorna -1.
        return current.val if current else -1

    def addAtHead(self, val: int) -> None:
        """
        Adiciona um novo nó com o valor especificado no início da lista.
        """

        # Cria um novo nó usando a própria classe MyLinkedList como estrutura do nó.
        new_node = MyLinkedList()

        # Atribui o valor ao atributo 'val' do novo nó.
        new_node.val = val

        # Define o próximo nó do novo nó como o nó 'head' atual (o antigo primeiro nó).
        new_node.next = self.head

        # Atualiza o 'head' da lista para apontar para o novo nó, tornando-o o primeiro nó.
        self.head = new_node

    def addAtTail(self, val: int) -> None:
        """
        Adiciona um novo nó com o valor especificado no final da lista.
        """

        # Cria um novo nó usando a própria classe MyLinkedList.
        new_node = MyLinkedList()

        # Atribui o valor ao atributo 'val' do novo nó.
        new_node.val = val

        # Se a lista estiver vazia (head é None), o novo nó se torna o 'head'.
        if not self.head:
            self.head = new_node
            return

        # Se a lista não estiver vazia, percorre a lista até o último nó.
        current = self.head
        while current.next:
            current = current.next

        # Define o próximo nó do último nó como o novo nó, adicionando-o ao final da lista.
        current.next = new_node

    def addAtIndex(self, index: int, val: int) -> None:
        """
        Adiciona um novo nó com o valor dado no índice dado.

        Args:
            index: O índice no qual adicionar o novo nó.
            val: O valor a ser adicionado à lista.
        """
        # Se o índice for negativo, chama a função addAtHead para adicionar no início da lista.
        if index < 0:
            self.addAtHead(val)
            return
        # Cria um novo nó com a classe MyLinkedList.
        new_node = MyLinkedList()
        # Atribui o valor ao atributo val do novo nó.
        new_node.val = val

        # Define a variável current como a cabeça da lista.
        current = self.head
        # Inicializa o contador count como 0.
        count = 0
        # Define a variável prev como None, que será usada para rastrear o nó anterior.

        prev = None

        # Percorre a lista até atingir o índice desejado ou o final da lista.
        while current and count < index:
            # Define prev como o nó atual.
            prev = current
            # Move current para o próximo nó.
            current = current.next
            # Incrementa o contador.
            count += 1

        # Se prev for None (ou seja, o índice é 0), chama a função addAtHead para adicionar o novo nó no início.
        if prev is None:
            self.addAtHead(val)
        # Se current for None (ou seja, o índice é maior ou igual ao tamanho da lista), chama a função addAtTail para adicionar o novo nó no final.
        elif current is None:
            self.addAtTail(val)
        # Caso contrário, insere o novo nó entre prev e current.
        else:
            # Define o próximo nó de prev como o novo nó.
            prev.next = new_node
            # Define o próximo nó do novo nó como current.
            new_node.next = current

    def deleteAtIndex(self, index: int) -> None:
        """
        Exclui o nó no índice dado.

        Args:
            index: O índice do nó a ser excluído.
        """
        # Se o índice for negativo, retorna sem fazer nada.
        if index < 0:
            return

        # Define a variável current como a cabeça da lista.
        current = self.head
        # Define a variável prev como None.
        prev = None

        # Se o índice for 0 (ou seja, excluindo a cabeça), atualiza a cabeça para o próximo nó (se houver) e retorna.
        if index == 0:
            self.head = current.next if current else None
            return

        # Inicializa o contador count como 0.
        count = 0
        # Percorre a lista até atingir o índice desejado ou o final da lista.
        while current and count < index:
            # Define prev como o nó atual.
            prev = current
            # Move current para o próximo nó.
            current = current.next
            # Incrementa o contador.
            count += 1

        # Se current não for None (ou seja, o nó a ser excluído existe), exclui o nó.
        if current:
            # Se prev não for None (ou seja, não estamos excluindo a cabeça), define o próximo nó de prev como o próximo nó de current.
            if prev:
                prev.next = current.next
            # Caso contrário (ou seja, estamos excluindo a cabeça), atualiza a cabeça para o próximo nó de current.
            else:
                self.head = current.next
````

````python
# Criando uma instância da lista ligada
myLinkedList = MyLinkedList()

# Chamando os métodos da lista ligada
print(myLinkedList.addAtHead(1))  # Output: null
print(myLinkedList.addAtTail(3))  # Output: null
print(myLinkedList.addAtIndex(1, 2))  # Output: null
print(myLinkedList.get(1))  # Output: 2
print(myLinkedList.deleteAtIndex(1))  # Output: null
print(myLinkedList.get(1))  # Output: 3
````

## Problem Description

Hardwoods are the botanical group of trees that have broad leaves, produce a fruit or nut, and generally go dormant in the winter. [cite: 1, 2, 3, 4, 5, 6, 7] Although oak, maple and cherry all are types of hardwood trees, for example, they are different species.

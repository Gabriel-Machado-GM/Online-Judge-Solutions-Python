## Problem Description

I have recently downloaded an iPhone app called NumPuzz, developed by Marmottajr. [cite: 1, 2] I do not read Portugese, but I find this little game quite suitable for our newbies contest. [cite: 2, 3] (The app is available for free at the time this problem description is written.) [cite: 3]

You are given a 3-by-3 square matrix with integer entries that may range from 0 to 9, and your task is to make all the entries zeros. [cite: 3, 4] You may click on a square and have that entry, together with its neighbouring entries, decreased by one. [cite: 4, 5] If an entry to be reduced is a zero, it becomes a 9 after the click. [cite: 5, 6] Of course, you aim at solving the puzzle with the fewest clicks possible. [cite: 6, 7]

As a simple example, let us suppose that your initial matrix has two zeros in the last two squares, and 1s everywhere else. [cite: 7, 8] If you click on the upper right corner, three of the ones turn into zeros. [cite: 8, 9] Now it takes just one more click to solve the puzzle. See the illustration below. [cite: 9, 10]

A player shows you his sequence of clicks to solve a NumPuzz instance (not necessarily optimal). [cite: 10, 11] Write a program that returns the initial configuration. [cite: 11]

## Input

Input contains no more than 100 lines. [cite: 11, 12] Each line gives a string of length not exceeding 200, describing the sequence of the player's moves. [cite: 12, 13] The meanings of the characters are as follows:

a b c
d e f
g h i


The solution discussed above is described by 'cd' (click first on c, then on d). [cite: 14, 15]

## Output

For each test case, print the corresponding matrix in the beginning of the game, using the format shown below. [cite: 15, 16] Notice that the second line in sample input is empty. [cite: 16, 17] It means that no clicks are required to solve that puzzle. [cite: 16, 17]

## Sample Input

cd\
\
cdifbgah


## Sample Output

Case #1:\
111\
111\
100\
Case #2:\
000\
000\
000\
Case #3:\
333\
343\
333

```python

test_no = 1
while True:
  try:
    seq = input()
    table = [0 for i in range(9)]

    # a0 b1 c2
    # d3 e4 f5
    # g6 h7 i8

    for ch in seq:
      if ch == "a":
        table[0] += 1
        table[1] += 1
        table[3] += 1
      elif ch == "b":
        table[0] += 1
        table[1] += 1
        table[2] += 1
        table[4] += 1
      elif ch == "c":
        table[1] += 1
        table[2] += 1
        table[5] += 1
      elif ch == "d":
        table[0] += 1
        table[3] += 1
        table[4] += 1
        table[6] += 1
      elif ch == "e":
        table[1] += 1
        table[3] += 1
        table[4] += 1
        table[5] += 1
        table[7] += 1
      elif ch == "f":
        table[2] += 1
        table[4] += 1
        table[5] += 1
        table[8] += 1
      elif ch == "g":
        table[3] += 1
        table[6] += 1
        table[7] += 1
      elif ch == "h":
        table[4] += 1
        table[6] += 1
        table[7] += 1
        table[8] += 1
      else:
        table[5] += 1
        table[7] += 1
        table[8] += 1

    table = [str(x%10) for x in table]
    print("Case #{}:".format(test_no))
    print("\n".join(
      [
        " ".join(table[0:3]),
        " ".join(table[3:6]),
        " ".join(table[6:9])
      ]
      ))

    test_no += 1
  except(EOFError):
    break
```
